```python
import numpy as np
import pandas as pd
```

```python
S0 = 400
U = 1.2
D = 0.9
R = 0.1
Q = (1+R-D)/(U-D)

original_cost = 200
lease_cost = 240

original_rate = 10000
enhance_rate = 14000

option_cost = 5000000
```

```python
def generate_gold_price_lattice(s, u, d):
    gold_lattice = pd.DataFrame(0,index=range(11),columns=range(11))
    gold_lattice.iloc[0,0] = s
    for i in range(1,11):
        for j in range(0,i+1):
            gold_lattice.iloc[j,i] = s * u**j * d**(i-j)
    return gold_lattice
```

```python
gold_price_lattice = generate_gold_price_lattice(S0, U, D)
gold_price_lattice
```

```python
def value_max(a, b):
    if a > b:
        return a 
    else:
        return b
```

```python
def generate_lease_value(gold_lattice, q, r, lease_c, rate):
    lease_value = pd.DataFrame(0,index=range(11),columns=range(11))
    for i in reversed(range(10)):
        for j in reversed(range(i+1)):
            lease_value.iloc[j,i] = (value_max(gold_lattice.iloc[j,i] - lease_c, 0)*rate + 
                                     q*lease_value.iloc[j+1, i+1] + (1-q)*lease_value.iloc[j, i+1]) / (1 + r)
    return lease_value
```

## Q1 Compute the time t=0 value of the mine when the enhancement is already in place.

```python
lease_value = generate_lease_value(gold_price_lattice, Q, R, lease_cost, enhance_rate)
lease_value
```

```python
original_value = generate_lease_value(gold_price_lattice, Q, R, original_cost, original_rate)
original_value
```

```python
def generate_option_value(gold_price, lease_v, option_c, original_c, original_r, q, r, t):
    option_value = pd.DataFrame(0,index=range(11),columns=range(11))
    exercise_boundary = pd.DataFrame(0,index=range(11),columns=range(11))
    
    for i in reversed(range(10)):
        for j in reversed(range(i+1)):
            if i < t:
                option_value.iloc[j, i] = (value_max(gold_price.iloc[j,i] - original_c, 0) * original_r + 
                                                    q * option_value.iloc[j+1, i+1] + (1 - q) * option_value.iloc[j, i+1]) / (1 + r)
            else:
                option_value.iloc[j, i] = value_max(lease_v.iloc[j, i] - option_c,
                                                    (value_max(gold_price.iloc[j,i] - original_c, 0) * original_r + 
                                                    q * option_value.iloc[j+1, i+1] + (1 - q) * option_value.iloc[j, i+1]) / (1 + r))
                exercise_boundary.iloc[j, i] = lease_v.iloc[j, i] - option_c - (value_max(gold_price.iloc[j,i] - original_c, 0) * original_r + 
                                                    q * option_value.iloc[j+1, i+1] + (1 - q) * option_value.iloc[j, i+1]) / (1 + r)
    return option_value, exercise_boundary
```

```python
option_value, exercise_boundary = generate_option_value(gold_price_lattice, lease_value, option_cost, original_cost, original_rate, Q, R, 5)
```
